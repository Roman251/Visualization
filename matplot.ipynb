```python
plt.style.use("fivethirtyeight")

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
#To explode any category, match it's index in the labels list with it's index in the explode list.
#Here 0.1 in the explode list corresponds to Python in the labels list.
plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([0,0.8,0.9,0.1,-0.8,-1])
plt.scatter(x,y)
p1 = np.polyfit(x,y,1)
print(p1)
#Returns an array that represents the equation of the curve used to fit the given data. The maximum coeff of X is 1.
#[-0.30285714  0.75714286] -> -0.30285714X+0.75714286
#Just like in ML, the polyfit function trains the data by considering a polynomial equation to fit the data with a given coeff of x.

print(np.polyval(p1,0.4))
#After p1 is trained using polyfit, np.polyval(p1,0.4) can be used to predict for any value(0.4 in this case).  
#This is similar to predicting a value for a data after training it.
# Another method for prediction is used below. 

p2 = np.polyfit(x,y,2)

#Returns an array that represents the equation of the curve used to fit the given data. The maximum coeff of X is 2.
#[-0.16071429  0.50071429  0.22142857] -> -0.16071429X^2+0.50071429X+0.22142857

```

```python
p3 = np.polyfit(x,y,3)
plt.plot(x,y,'o')
xp = np.linspace(-2,6,100)
plt.plot(xp,np.polyval(p1,xp),'r-')
plt.plot(xp,np.polyval(p2,xp),'b--')
plt.plot(xp,np.polyval(p3,xp),'m:')
```

```python
np.polyval(p3,0.4)
```

```python
df = pd.read_csv('pol.csv')
X=df.iloc[:, 1:2].values
y=df.iloc[:, 2].values
```

```python
X
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly

'''
For a data with only one independent variable, the polynomial equation derived for degeree=2 would be:
y = w0+w1x+w2x^2
For a data with 2 independent variable, the polynomial equation derived for degeree=2 would be:
y = X*X1+X^0*X1^2+X^2*X1^0+X^1*X1^0+X^0*X1^1+X^0*X1^0

'''
```

```python
poly_reg.fit(X_poly,y)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
#linear
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#Polynomial
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color='blue')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
